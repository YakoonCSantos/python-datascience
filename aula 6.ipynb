```python
titles_sliced = titles_basic.iloc[:,[0,1,2,8]]  #minimizar a tabela para uma visualiza√ß√£o curta, com
titles_sliced.head(5)
```

```python
titles_sliced.columns = ['tconst','tipo', 'titulo', 'genero']  ###renomear o nome das colunas com a fun√ß√£o Columns 
titles_sliced.head(5)
```

```python
titles_rating_merge = titles_rating.merge(titles_sliced, left_on = 'tconst', right_on = 'tconst') # fazendo a jun√ß√£o das tabelas com Merge, utilizando a coluna TCONST como chave primaria   
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.merge.html?highlight=merge#pandas.merge
titles_rating_merge.head(5)
```

```python
titles_rating_merge.sort_values(by =['numVotes'], ascending=False).head(100) # fa√ßa a ordena√ß√£o da coluna NUMVOTES do maior para o menor e imprima as 30 primeiras linhas
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_values.html?highlight=sort_values
```

```python
top = titles_rating_merge[(titles_rating_merge['numVotes'] > 50000) & (titles_rating_merge['tipo'] == 'movie')].sort_values(by =['averageRating'], ascending=False).head(50)
top
```

```python
top1 = titles_rating_merge[(titles_rating_merge['tipo'] == 'movie')].sort_values(by =['numVotes'], ascending=False).head(50)
top1
```

```python
top2 = top.groupby(['tipo', 'genero']).agg({'numVotes':'sum'}) #contru√ß√£o da tabela dinamica , pivotagem
top2
```

```python
top3 = top.groupby(['tipo','titulo']).agg({'numVotes':'sum'})
top3
```

```python
top_pcts = top3.groupby(level = 0).apply(lambda x: 100 * x / float(x.sum()))
top_pcts.sort_values(by = ['numVotes'], ascending = False)
```

```python
titles_sliced = titles_basic.iloc[:,[0,1,2,8]]  #minimizar a tabela para uma visualiza√ß√£o curta, com
genres = titles_sliced['genres'].str.split(',', 1, expand = True )
#generos.head()
titles_sliced['gerero_1'] = genres[0]
titles_sliced['gerero_2'] = genres[1]
titles_sliced.drop(columns = ['genres'], inplace = True)
titles_sliced.head(5)
```

```python
titles_sliced.columns = ['tconst', 'tipo', 'titulo', 'genero', 'sub.genero']
titles_sliced.head()
```

```python
titles_rating_merge = titles_rating.merge(titles_sliced, left_on = 'tconst', right_on = 'tconst') # fazendo a jun√ß√£o das tabelas com Merge, utilizando a coluna TCONST como chave primaria   
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.merge.html?highlight=merge#pandas.merge
titles_rating_merge.head()
```

```python
titles_rating_merge.sort_values(by =['numVotes'], ascending=False).head(100)
```

```python
top = titles_rating_merge[(titles_rating_merge['numVotes'] > 50000)].sort_values(by =['averageRating'], ascending=False).head(50)
top
```

```python
top_1 = top.groupby(['tipo','genero']).agg({'numVotes':'sum'})
top_1
```

```python
top_pcts = top_1.groupby(level = 0).apply(lambda x: 100 * x / float(x.sum()))
top_pcts.sort_values(by = ['numVotes'], ascending = False)
```

```python
top_2 = top.groupby(['tipo','genero']).agg({'averageRating':'mean', 'numVotes':'sum'})
top_2
```
